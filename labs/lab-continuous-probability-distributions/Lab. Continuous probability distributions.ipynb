```python
def generate_uniform(bottom, ceiling, count):
    return uniform.rvs(loc = bottom, scale = ceiling - bottom, size = count)
```

```python
uni1 = generate_uniform(10, 15, 100)
uni1
```

```python
uni2 = generate_uniform(10, 60, 1000)
uni2
```

```python
plt.hist(uni1, bins = 10)
plt.show()
```

```python
plt.hist(uni2, bins = 10)
```

```python
from scipy.stats import norm 
```

Which one(s) of the variables are nearly normally distributed? How do you know?

```python
#From the histograms, we can clearly see that the data is not normally distributed, because the graphs are not symmetrical.
```

```python
#let's check some stats...
import scipy.stats as stats
print(stats.describe(vehicles['Fuel Barrels/Year']))
print(stats.describe(vehicles['CO2 Emission Grams/Mile']))
print(stats.describe(vehicles['Combined MPG']))

#The skewness is positive for all three columns, which means that the data is skewed right, with Combined MPG being the most skewed.
#From the kurtosis values, Fuel Barrels/Year and CO2 Emission Grams/Mile are less peaked than a normal distribution (platykurtic), while Combined MPG is closer to a normal distribution but still not quite there.
```

```python
# let's do some chat-gpt normality tests :D
def normality_test(data):
    # Shapiro-Wilk test
    shapiro_test = stats.shapiro(data)
    
    # Kolmogorov-Smirnov test
    ks_test = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
    
    # Anderson-Darling test
    anderson_test = stats.anderson(data, dist='norm')
    print(f'Shapiro-Wilk test: Statistic={shapiro_test[0]}, p-value={shapiro_test[1]}')
    print(f'Kolmogorov-Smirnov test: Statistic={ks_test[0]}, p-value={ks_test[1]}')
    print(f'Anderson-Darling test: Statistic={anderson_test.statistic}, Critical values={anderson_test.critical_values}')

print('Fuel Barrels/Year')
normality_test(vehicles['Fuel Barrels/Year'])
print('CO2 Emission Grams/Mile')
normality_test(vehicles['CO2 Emission Grams/Mile'])
print('Combined MPG')
normality_test(vehicles['Combined MPG'])
```

None of them are normally ditributed. 

## Some chat gpt Conclusions:

### Shapiro-Wilk Test
The Shapiro-Wilk test is sensitive to departures from normality.
For all variables, the p-values are extremely low (far below 0.05), indicating that none of the distributions are normal.

### Kolmogorov-Smirnov Test
The Kolmogorov-Smirnov test compares the sample distribution to a normal distribution.
The p-values are also zero for all variables, indicating significant departures from normality.

### Anderson-Darling Test
The Anderson-Darling test provides a test statistic which, if greater than the critical values, suggests non-normality.
The test statistics for all variables are significantly higher than the critical values, reinforcing the results of the Shapiro-Wilk and Kolmogorov-Smirnov tests.

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

```python
def exponential_distribution_list(count, avg=10):
    return np.random.exponential(avg, count)
```

```python
plt.hist(exponential_distribution_list(10), bins=100)
```

```python
plt.hist(exponential_distribution_list(100), bins=100)
```

How are the two distributions different?

The second histogram, with 100 data points, clearly shows the exponential decay pattern, while the first histogram, with only 10 data points, is too sparse and irregular to reveal the distribution's characteristics effectively.

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
lambda_value = 1 / 10
time = 15 

p = 1 - np.exp(-lambda_value * time)

p
```
