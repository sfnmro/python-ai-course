## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
from math import comb

white_balls = 10
red_balls = 12
black_balls = 8
total_balls = white_balls + red_balls + black_balls

# Probability of taking a white ball out
p_white = white_balls / total_balls
print(f"P(white ball) = {white_balls}/{total_balls} = {p_white:.4f}")

# Probability of taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
p_white_then_black_no_replace = (white_balls / total_balls) * (black_balls / (total_balls - 1))
print(f"P(white ball on first extraction and black ball on second without replacement) = {p_white_then_black_no_replace:.4f}")

# Probability of taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
p_white_then_black_replace = (white_balls / total_balls) * (black_balls / total_balls)
print(f"P(white ball on first extraction and black ball on second with replacement) = {p_white_then_black_replace:.4f}")

# Probability of taking 3 balls with all different colors
p_three_different_colors = white_balls * red_balls * black_balls / comb(total_balls, 3)
print(f"P(taking 3 balls with all different colors) = {p_three_different_colors:.4f}")
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
"""
your solution here


We want to know P(R)


P(R) = P(C & R) / P(not C & R)
       = (0.4*0.5) + (0.6*0.2)
       = 0.32
       

"""

p_cloudy = 0.40
p_rain_given_cloudy = 0.50
p_rain_given_not_cloudy = 0.20

# Probability of a non-cloudy day
p_not_cloudy = 1 - p_cloudy

# Using the law of total probability to find the probability of rain
p_rain = (p_cloudy * p_rain_given_cloudy) + (p_not_cloudy * p_rain_given_not_cloudy)

p_rain
```

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Given probabilities
P_A = 0.3
P_B = 0.6
P_A_and_B = 0.1

# Create a Venn diagram
plt.figure(figsize=(8, 8))
venn = venn2(subsets=(P_A - P_A_and_B, P_B - P_A_and_B, P_A_and_B), set_labels=('A', 'B'))

# Set the labels for each region
venn.get_label_by_id('10').set_text(f'A\n{P_A - P_A_and_B}')
venn.get_label_by_id('01').set_text(f'B\n{P_B - P_A_and_B}')
venn.get_label_by_id('11').set_text(f'A ∩ B\n{P_A_and_B}')

# Set title
plt.title("Venn Diagram for Events A and B")
plt.show()

# Given probabilities
P_A = 0.3
P_B = 0.6
P_A_and_B = 0.1

# Calculate derived probabilities
P_A_complement = 1 - P_A
P_A_union_B = P_A + P_B - P_A_and_B
P_A_intersect_B_complement = P_A - P_A_and_B
P_A_union_A_complement = 1  # Always true
P_A_intersect_B_union_B_complement = P_A_and_B_complement = 0.4  # Corrected as true
P_A_intersect_B = P_A_and_B
P_A_complement_intersect_B_complement = 1 - P_A_union_B

# Create a dictionary with the results
results = {
    "P(A')": P_A_complement,
    "P(A U B)": P_A_union_B,
    "P(A ∩ B')": P_A_intersect_B_complement,
    "P(A U A')": P_A_union_A_complement,
    "P(A ∩ (B U B'))": P_A_intersect_B_union_B_complement,
    "P(A ∩ B)": P_A_intersect_B,
    "P(A' ∩ B')": P_A_complement_intersect_B_complement
}

print("Results:")
for key, value in results.items():
    print(f"{key} = {value:.1f}")
```

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

Sum of 7: There are 6 combinations (1+6, 2+5, 3+4, 4+3, 5+2, 6+1) out of 36 possible outcomes, giving a probability of 6/36 = 1/6 ≈ 0.167.

Sum of 8: There are 5 combinations (2+6, 3+5, 4+4, 5+3, 6+2) out of 36 possible outcomes, giving a probability of 5/36 ≈ 0.139.

Sum of 9: There are 4 combinations (3+6, 4+5, 5+4, 6+3) out of 36 possible outcomes, giving a probability of 4/36 = 1/9 ≈ 0.111.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Probability calculations for the sums
def probability_sum(dice_sides, target_sum):
    count = 0
    total_outcomes = dice_sides ** 2
    for i in range(1, dice_sides + 1):
        for j in range(1, dice_sides + 1):
            if i + j == target_sum:
                count += 1
    return count / total_outcomes

# Calculate probabilities
dice_sides = 6
prob_7 = probability_sum(dice_sides, 7)
prob_8 = probability_sum(dice_sides, 8)
prob_9 = probability_sum(dice_sides, 9)

# Print the calculated probabilities
print(f"P(sum=7) = {prob_7:.4f}")
print(f"P(sum=8) = {prob_8:.4f}")
print(f"P(sum=9) = {prob_9:.4f}")


# Simulation of rolling 2 dice 10,000 times
n_simulations = 10000
dice_rolls = np.random.randint(1, 7, size=(n_simulations, 2))
sums = np.sum(dice_rolls, axis=1)

# Plot the count plot
plt.figure(figsize=(10, 6))
sns.countplot(x=sums)
plt.title('Count Plot of Sum of 2 Dice Rolls (10,000 simulations)')
plt.xlabel('Sum')
plt.ylabel('Count')
plt.show()
```
